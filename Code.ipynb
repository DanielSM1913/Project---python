```python
## Importing libraries and defining variables

import pandas as pd
import matplotlib.pyplot as plt
cyber_thr = pd.read_csv ('Global_Cybersecurity_Threats_2015-2024.csv')

```

```python
###Testing Database

print(cyber_thr.columns)
print(cyber_thr.describe())

## First line of code are made to test,verify and see general info about any database in this case cyber-attacks database.
## we define a varibel to call the entire database in this case "cyber_thr"
```

```python
## Question. Incident Resolution Time Average for common attacks.


ir_co = cyber_thr.groupby("Attack Type")["Incident Resolution Time (in Hours)"].mean().reset_index()

plt.figure(figsize=(8, 3))
plt.bar(ir_co["Attack Type"], ir_co["Incident Resolution Time (in Hours)"].astype(int))

plt.xlabel("Attack_Type")
plt.ylabel("Resolution_Time_Hours")
plt.title("Incident Resolution Time Average for common attacks")
print(ir_co)

##The first question we tried to answer is How much the IT personal takes to resolve an incident in average.

## First we started defining a variable which who will use the variable made for the database.
## Then we used the groupby command to put together 2 different section of a data base.
## In this case we used "Attack Type", and "Incident Resolution Time (in Hours)"
## Then We used the command astype to put the values as an integer and not as a string because we are using numbers.
## Finally we used some basic command to build a graph, setting names for the axies, and title
## At the end we call the entire variable as an extra to see specif values.
```

```python
## Question. Most Targeted Industries.

ind_cou = (cyber_thr['Target Industry'].value_counts())
plt.figure(figsize=(15, 8))

plt.bar(ind_cou.index, ind_cou.values)

plt.xlabel("Industries")
plt.ylabel("Number of Attacks")
plt.title("Number of Attacks per Industry")

print(ind_cou)


# The second question we wanted to answer was which industries are the most targeted.

## This was a fast coding problem,we just defined a varibel for the "Targeted industries"
##Then we just used the value_counts command to take the times that indutry were targeted
## FInally we just build the graph defing, and putting the names for each axies
## Again we call the function at then end as an extra to see specific values.
```

```python
## Question. Most Common Defense Mechanism Used.

defend_mech = cyber_thr['Defense Mechanism Used'].value_counts()

plt.figure(figsize=(10, 8))
plt.bar(defend_mech.index, defend_mech.values)
plt.xlabel("Defense Mechanism Used")
plt.ylabel("Number of times Used")
plt.title("Defense Mechanism Used")
print(defend_mech)

## The question for this code was to find which defense mechanism were used against these incidents.

##This was a very similar challenge as previous question, we just defined a variable for the "Defense Mechanism Used"
## Then we just added the values_counts command to count every time each mechanism were used
## Finally we just created the bar chart, defining every axies, also we used .index and .values so the graph can identify the values for each specific mechanism type.
```

```python
## Relation between Financial loss and Attack source

Fin_Att = cyber_thr.groupby('Attack Type')['Financial Loss (in Million $)'].mean()

plt.figure(figsize=(10, 6))
plt.bar(Fin_Att.index, Fin_Att.values)

plt.xlabel("Attack Type")
plt.ylabel("Average Financial Loss (Million $)")
plt.title("Average Financial Loss per Attack Type")

print(Fin_Att)



## The question we answer this time is the relation between the financial loss and the type of attack
## In this case we defined a variable and calling the databse with groupby command to take two different section of the database
## Then we used the command .mean to the financil loss part becasue we want the average values for this section only
##Finally we build the graph and caliing the .index and .values to get the right string and average values for each attack type.
## Finally we called the whole variable to verigy and see the specific values for each attack type.
```
